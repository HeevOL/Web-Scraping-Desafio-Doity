```python
# Importes necessarios para fazer o scraping.
import requests
from bs4 import BeautifulSoup
# import json
import pandas as pd
# from pandas import json_normalize
# import numpy as np
```

```python
# Função para converter o nome do mês para número.
def mes(string):
    if string == 'janeiro': return '1'
    elif string == 'fevereiro': return '2'
    elif string == 'março': return '3'
    elif string == 'abril': return '4'
    elif string == 'maio': return '5'
    elif string == 'junho': return '6'
    elif string == 'julho': return '7'
    elif string == 'agosto': return '8'
    elif string == 'setembro': return '9'
    elif string == 'outubro': return '10'
    elif string == 'novembro': return '11'
    else: return '12'
```

```python
def scraping(pag):
    # headers = {'user-agent': 'Mozilla/5.0'} # Caso o site identifique como um rôbo.
    # Página web salva com o request.
    pagina = requests.get(pag) 

    # Aqui é salvo o conteudo html da página de forma desorganizada.
    soup = pagina.text 

    # Aqui é salvo o conteudo estruturado da página.
    strutured_soup = BeautifulSoup(soup, 'html.parser')

    # Encontro na estrutura HTML as informações de nome do evento e data/horario de inicio e encerramento.
    sobre_evento = strutured_soup.find('div',{'class':'col-md-9'})

    # Salva o nome do Evento.
    nome = sobre_evento.contents[1].string

    # Salva a data.
    data = sobre_evento.find('div',{'class':'custom-data'}).contents[2].string.strip()
    data = data.replace("de", "").replace(",","").split()
    inicio = data[0]+'/'+mes(data[1])+'/'+data[2]+' '+data[3][:2]+':'+data[3][3:]
    if data[4] == '-':
        termino = data[0]+'/'+mes(data[1])+'/'+data[2]+' '+data[5][:2]+':'+data[5][3:]
    else:
        termino = data[5]+'/'+mes(data[6])+'/'+data[7]+' '+data[8][:2]+':'+data[8][3:]

    # Salva a modalidade e local caso seja presencial.
    modalidade = strutured_soup.find('div', {'class':"tooltip-mapa"}).contents[0].strip()

    if 'online'.lower() in modalidade.lower():
        modalidade = "Online"
        local = "Web"
    else:
        modalidade = strutured_soup.find('div', {'class':'custom-local'}).contents[2].string.strip()
        local = modalidade
        modalidade = "Presencial"

    # Salva o organizador do evento.
    organizador = strutured_soup.find("div", {"class":"col-md-12"}).contents[1].string

    # Salva os dados encontrados na pagina em um dicionario.
    dados ={
        "Nome do Evento": nome,
        "Início": inicio,
        "Termino": termino,
        "Modalidade": modalidade,
        "Local": local,
        "Organizador": organizador
        }
    # Retorna o dicionario dados.
    return dados    
```

```python
# main
# Aqui criei um data frame vazio que será iterado com os resultados da lista pagina, pela função Scraping().
df = pd.DataFrame()
paginas = [
    'https://doity.com.br/desafionordeste420km',
    'https://doity.com.br/corridacaapi',
    'https://doity.com.br/aditinvest2022',
    'https://doity.com.br/rogadx',
    'https://doity.com.br/colposcopia2022',
    'https://doity.com.br/missao-asia',
    'https://doity.com.br/congresso-nortenordeste-e-centro-oeste-de-cirurgia-torcica',
    'https://doity.com.br/2-curso-intensivo-de-nutrio-clnica-avanada--vsf',    
    'https://doity.com.br/diadoacbr2022',
    'https://doity.com.br/cnma2022',
    'https://doity.com.br/isimposiopernambucanodeatendimentoprehospitalar',
    'https://doity.com.br/onco2022',
    'https://doity.com.br/corrida-pet-run-teresina-piaui',
    'https://doity.com.br/folegoparaeuropa-2022',
    'https://doity.com.br/corridasunimed2022',
    'https://doity.com.br/complan2022',
    'https://doity.com.br/cinasama-congresso-internacional-de-saude-e-meio-ambiente-iii-edicao-on-line-2022',
    # 'https://doity.com.br/primeira-corrida-de-rua-37-anos-faculdade-cesvale', Página removida
    'https://doity.com.br/deltasaude2022',
    'https://doity.com.br/festquali-parana-2022',
    'https://doity.com.br/congresso-aeromedico-brasileiro-2022',
    'https://doity.com.br/missaotecnicabairrosplanejadosdepelotas',
    'https://doity.com.br/immersion-training-2022',
    'https://doity.com.br/iii-simposio-do-ruminacao',
    'https://doity.com.br/desafiodelta2022'
]
# Esse for adiciona os dados extraidos de cada uma das páginas para o data frame pandas.
for url in paginas:
    df = df.append(scraping(url), ignore_index=True)
```

```python
# Transforma o data frame em uma tabela Excel.
df.to_excel('web-scraping-doity.xlsx', index_label=False)
```
